```python
import pickle

X_train_A = pickle.load(open('X_train_A.pkl', 'rb'))
X_train_B = pickle.load(open('X_train_B.pkl', 'rb'))
X_test_A = pickle.load(open('X_test_A.pkl', 'rb'))
X_test_B = pickle.load(open('X_test_B.pkl', 'rb'))

y_train_A = pickle.load(open('y_train_A.pkl', 'rb'))
y_train_B = pickle.load(open('y_train_B.pkl', 'rb'))
y_test_A = pickle.load(open('y_test_A.pkl', 'rb'))
y_test_B = pickle.load(open('y_test_B.pkl', 'rb'))
```

```python
from sklearn.feature_extraction.text import CountVectorizer

X_train_A = X_train_A.fillna('')
X_train_B = X_train_B.fillna('')
X_test_A = X_test_A.fillna('')
X_test_B = X_test_B.fillna('')

vectorizer = CountVectorizer()
XA = vectorizer.fit_transform(X_train_A)
YA = y_train_A

XA_test = vectorizer.transform(X_test_A)
YA_test = y_test_A

XB = vectorizer.fit_transform(X_train_B)
YB = y_train_B
XB_test = vectorizer.transform(X_test_B)
YB_test = y_test_B

```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier

#  Base A
neigh = KNeighborsClassifier(n_neighbors=8)
neigh.fit(XA, YA)

KNNPredictionsA = neigh.predict(XA_test)


# Base B
neigh = KNeighborsClassifier(n_neighbors=2)
neigh.fit(XB, YB)

KNNPredictionsB = neigh.predict(XB_test)
```

```python
from sklearn.linear_model import LogisticRegression

# Base A
model = LogisticRegression(random_state=42, max_iter=200, C=1.0, penalty='l2', solver='newton-cg')
model.fit(XA, YA)

logRegPredictionsA = model.predict(XA_test)

# Base B
model = LogisticRegression(random_state=42, max_iter=200, C=100.0, penalty='l2', solver='lbfgs')
model.fit(XB, YB)

logRegPredictionsB = model.predict(XB_test)
```

```python
from sklearn.naive_bayes import MultinomialNB

# Base A
mnb = MultinomialNB(alpha=1.2, fit_prior=True, force_alpha=True)
mnb.fit(XA, YA)

naivePredictionsA = mnb.predict(XA_test)

# Base B
mnb = MultinomialNB(alpha=0.3, fit_prior=True, force_alpha=True)
mnb.fit(XB, YB)

naivePredictionsB = mnb.predict(XB_test)
```

```python
from sklearn import metrics

def printMetrics(y_test, predictions):
    print("accuracy: ", metrics.accuracy_score(y_test, predictions))
    print("f1_score micro: ", metrics.f1_score(y_test, predictions, average='micro'))
    print("f1_score macro: ", metrics.f1_score(y_test, predictions, average='macro'))

    # print("Confusion matrix\n", metrics.confusion_matrix(y_test, predictions))
```

```python
print("base A")
print("3 Nearest Neighbors:")
printMetrics(YA_test, KNNPredictionsA)

print("\nNaive Bayes:")
printMetrics(YA_test, naivePredictionsA)

print("\nLogistical Regression:")
printMetrics(YA_test, logRegPredictionsA)

print("base B")
print("3 Nearest Neighbors:")
printMetrics(YB_test, KNNPredictionsB)

print("\nNaive Bayes:")
printMetrics(YB_test, naivePredictionsB)

print("\nLogistical Regression:")
printMetrics(YB_test, logRegPredictionsB)
```
